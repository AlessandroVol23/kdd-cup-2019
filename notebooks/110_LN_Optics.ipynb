```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import matplotlib.gridspec as gridspec
import seaborn as sns
from sklearn.cluster import OPTICS, cluster_optics_dbscan
import pandas as pd
import numpy as np
```

```python
df_profiles = pd.read_csv('../data/raw/profiles.csv')
df_profiles
```

```python
clust = OPTICS(min_samples=50, xi=.05, min_cluster_size=.05)

# Run the fit
clust.fit(df_profiles)

space = np.arange(len(df_profiles))
reachability = clust.reachability_[clust.ordering_]
labels = clust.labels_[clust.ordering_]
```

```python
len(labels)
```

```python
df_optics = df_profiles
df_optics['cluster_id'] = np.nan
df_optics
```

```python
for each in range(len(df_optics)):
    x = labels[each]
    df_optics.loc[df_optics.index == each, 'cluster_id'] = str(x)
    
df_optics
```

```python
n_cluster = df_optics.cluster_id.unique()
n_cluster
```

```python
cluster = []

for each in n_cluster:
    x = str(each)
    a = np.array(df_optics.pid.loc[df_optics['cluster_id'] == x])
    cluster.append(a)
cluster
```

```python
cluster[0]
```

```python
np.save('../data/task2/optics/optics.npy', labels)
np.save('../data/task2/optics/optics_cluster.npy', cluster)
```
