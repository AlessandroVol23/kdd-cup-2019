```python
import pandas as pd
```

```python
# Count how often one pid takes which click_mode
df_count_pid = df_train[['pid', 'click_mode']]
```

```python
def one_hot_encode(df_all, column_to_encode, target_column="click_mode"):
    # One hot encode month 
    one_hot = pd.get_dummies(df_all[column_to_encode])
    one_hot.columns = [column_to_encode + "," + str(col) for col in one_hot.columns]
    df_all = df_all.drop(columns=column_to_encode, axis=0)
    df_all = df_all.join(one_hot)
    if target_column is None:
        return df_all
    y = df_all[target_column]
    df_all = df_all.drop(columns=target_column, axis=0)
    df_all = df_all.join(y)
    return df_all
```

```python
df_count_pid = one_hot_encode(df_count_pid, 'click_mode', None)
```

```python
df_count_pid.head()
```

```python
df_count_pid=df_count_pid.groupby('pid').sum()
```

```python
df_train_merged_no_def = pd.merge(df_train, df_count_pid, on='pid')
```

```python
df_count_pid_min_1 = df_count_pid.copy()
```

```python
df_count_pid_min_1.loc[df_count_pid_min_1.index == -1] =-1
```

```python
df_train_merged_no_def_min_1 = pd.merge(df_train, df_count_pid_min_1, on='pid')
```
