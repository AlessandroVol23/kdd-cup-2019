```python
import pandas as pd
import numpy as np
#import matplotlip.pyplot as plt

df_train = pd.read_pickle("../data/external/districts/train_all_first_districts.pickle")
df_train
```

```python
beijing_districts = pd.read_csv("../data/external/districts/beijing_districts.csv")
beijing_districts
```

```python
# Euclidean Distance Caculator
def dist(a, b, ax=0):
    return np.linalg.norm(a - b, axis=ax)
```

```python
# Getting coordinates of Starting Points
f1 = df_train['o_lat'].values
f2 = df_train['o_long'].values
P = np.array(list(zip(f1, f2)))
len(P)
```

```python
array = df_train['o_district'].values
split_array = []
districts = []

for each in array:
    split = each.split(', ')
    split_array.append(split)

for nr in range(len(split_array)):
    z = split_array[nr][0].split('o_district_')
    
    if z[1] not in districts:
        districts.append(z[1])

districts.sort()
len(districts)
```

```python
districts
```

```python
for district in districts:
    df_train.loc[:,'o_distance_' + district] = np.nan
list(df_train.columns)
```

```python
df_train
```

```python
for p in range(len(P)):
    
    if p % 100 == 0:
        print("Processing row {}".format(str(p)), end="\r")
    
    for i in districts:
        i = int(i)
        x1 = beijing_districts['o_lat'].values[i]
        y1 = beijing_districts['o_long'].values[i]
        C = np.array([x1, y1])
        
        distance = dist(P[p], C)
        df_train.loc[p,'o_distance_' + str(i)] = distance
print("Calculate distances to districts successful")
```
