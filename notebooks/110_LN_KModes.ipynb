```python
import pandas as pd
import numpy as np
from kmodes.kmodes import KModes
```

```python
df_profiles = pd.read_csv('../data/raw/profiles.csv')
df_profiles
```

```python
km = KModes(n_clusters=4, init='Huang', n_init=5, verbose=1)
clusters = km.fit_predict(df_profiles)
```

```python
# Print the cluster centroids
km.cluster_centroids_
```

```python
labels = km.labels_
labels
```

```python
df_kmodes = df_profiles
cluster=[4, 8, 16, 36]
df_kmodes = df_kmodes.drop(['cluster[4, 8, 16, 36]_id'], axis=1)

for each in cluster:
    df_kmodes[f'cluster{each}_id']=np.nan
df_kmodes
```

```python
df_p = pd.read_csv('../data/raw/profiles.csv')

km8 = KModes(n_clusters=8, init='Huang', n_init=5, verbose=1)
clusters = km8.fit_predict(df_p)
labels8 = km8.labels_
labels8
```

```python
df_p = pd.read_csv('../data/raw/profiles.csv')

km16 = KModes(n_clusters=16, init='Huang', n_init=5, verbose=1)
clusters = km16.fit_predict(df_p)
labels16 = km16.labels_
labels16
```

```python
df_p = pd.read_csv('../data/raw/profiles.csv')

km36 = KModes(n_clusters=36, init='Huang', n_init=5, verbose=1)
clusters = km36.fit_predict(df_p)
labels36 = km36.labels_
labels36
```

```python
np.save('../data/task2/kmeans/hamming/kmode4.npy', labels)
np.save('../data/task2/kmeans/hamming/kmode8.npy', labels8)
np.save('../data/task2/kmeans/hamming/kmode16.npy', labels16)
np.save('../data/task2/kmeans/hamming/kmode36.npy', labels36)
```

```python
for each in range(len(df_kmodes)):
    x = labels[each]
    y = labels8[each]
    z = labels16[each]
    w = labels36[each]

    df_kmodes.loc[df_kmodes.index == each, 'cluster4_id'] = str(x)
    df_kmodes.loc[df_kmodes.index == each, 'cluster8_id'] = str(y)
    df_kmodes.loc[df_kmodes.index == each, 'cluster16_id'] = str(z)
    df_kmodes.loc[df_kmodes.index == each, 'cluster36_id'] = str(w)

df_kmodes
```

```python
df_kmodes.to_csv("../data/task2/kmeans/hamming/df_kmodes.csv")
```

```python
n_cluster = df_kmodes.cluster36_id.unique()
cluster36 = []

for each in n_cluster:
    x = str(each)
    a = np.array(df_kmodes.pid.loc[df_kmodes['cluster36_id'] == x])
    cluster36.append(a)
len(cluster36)
```

```python
np.save('../data/task2/cluster36.npy', cluster36)
```
